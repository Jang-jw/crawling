```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
head={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
```

```python
res=req.get('https://www.acmicpc.net/problemset',headers=head)
res
```

```python
res.text
```

```python
soup=bs(res.text,'lxml')
soup
```

```python
result=soup.select('#problemset > tbody > tr > td:nth-child(2) > a')
result
```

```python
len(result)
```

```python
result[0].text
```

```python
list_index=[]
list_title=[]
for i in range(len(result)):
    list_index.append(i+1)
    list_title.append(result[i].text)
```

```python
QuestionTitle={'번호':list_index,'제목':list_title}
resultP=pd.DataFrame(QuestionTitle)
resultP
```

```python
resultP.set_index('번호',inplace=True)
resultP
```

```python
resultP.to_csv('백준문제리스트1.csv',encoding='euc=kr')
```
