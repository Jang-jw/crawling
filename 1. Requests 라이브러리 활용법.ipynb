# Beautifulsoup 라이브러리
- req 로 넘겨받은 데이터(텍스트)를 컴퓨터가 이해할 수 있게 html 형식으로 변형 
- beautifulsoup 객체화

```python
from bs4 import BeautifulSoup as bs
```

```python
# bs('변형할 텍스트 데이터',"lxml")
# 텍스트 데이터를 컴퓨터가 이해할 수 있는 형태로 변형 > ''사라짐 
soup=bs(res.text,'lxml')

# soup 데이터에서 특정 '요소'를 수집
# soup.select('요소의 구분자') > 특정 요소만 수집 
# select vs selectone = 데이터의 수집 개수차이 (복수, 단수) 
news_title=soup.select('a.news_tit')
```

```python
soup.select_one('a.news_tit')
```

```python
news_title[0].text
```

```python
# 모든 뉴스의 타이틀을 출력
for i in news_title:
    print(i.text)
```

```python
text_title=[]
for i in news_title:
    text_title.append(i.text)
text_title
```

# 상위 증권종목 수집

```python
res=req.get('https://finance.naver.com/')
res.text
```

```python
soup=bs(res.text,'lxml')
```

```python
soup.select('#_topItems1>tr>th>a')
```

```python
soup.select('#_topItems1>tr>th>a')[0]
```

```python
soup.select('#_topItems1>tr>th>a')[0].text
```

```python
listStockTop=[]
for i in soup.select('#_topItems1>tr>th>a'):
    listStockTop.append(i.text)
listStockTop
```
